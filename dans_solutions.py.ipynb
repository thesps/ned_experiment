# Part 1

Need to use a prescale such that the deadtime is kept to a minimum - 3 is good.
To start with, look at the 2D distribution to get a feeling for the signal.

```python
data = DAQ.run({'energy_thresh': 50}, prescale = 3)
run.plot_data(data)
```

So, we expect occupancy to offer better discrimination using a threshold cut.
ROC curves are very useful to judge different discriminators in quantifiable way
(although as a function of data rate is just as useful).
Use the scan function for both variables. 

```python
tvars = ['energy_thresh', 'occupancy_thresh']
ROC_results = {}
for tvar in tvars:
    ROC_results[tvar] = run.scan_trigger(tvar, 3)
```

```python
plt.figure(figsize = [8, 6])
print(ROC_results.keys())
for tvar in tvars:
    plt.plot(ROC_results[tvar]['fprs'], ROC_results[tvar]['tprs'], label = tvar)
    #plt.plot(ROC_results[tvar]['storage_space_bytes'], ROC_results[tvar]['tprs'], label = tvar)
    
plt.xlabel('fpr')
plt.ylabel('tpr')
plt.grid()
plt.legend(loc = 'lower right')
plt.show()
```

Occupancy clearly gives superior discrimination. 
BUT, trigger discrimination isn't everything. If we trigger on occupancy,
we may well recover more energy events for a fixed data rate, 
but at the expense of introducing a bias at the trigger level.
If occupancy and energy wern't biased, it might be simpler, but 
if we have a bias, we have to rely on simulations or other runs to measure it.
--> less ideal from an analysis point of view. Might be preferable to 
trigger on energy, or trigger on a really low occupancy but with a pre-scale.

Not enough information to propose real settings - need to understand systematics
incurred by trigger bias on occupancy. Unless thats known, need to trigger
on energy, maybe with a pre-scale to get whole spectrum at a sustainble
rate with low deadtime.

# Part 2

High deadtimes can often introduce bias at the analysis level.
Deadtime is often caused by buffer overflow, or too much data.
Therefore, it might depend on the event types themselves - a high
occupancy event could lead to increased deadtime, so we may get a 
lack of high occupancy events in the output. These kinds of biases
are notoriously difficult to account for, and best avoided by keeping
deadtime to an absolute minimum.

```python
# Deadtime scan (same as above, but with no prescale).

tvars = ['energy_thresh', 'occupancy_thresh']
ROC_results = {}
for tvar in tvars:
    ROC_results[tvar] = run.scan_trigger(tvar, 1)
```

```python
plt.figure(figsize = [13, 6])
print(ROC_results.keys())
for itvar, tvar in enumerate(tvars):
    plt.subplot(1,2,itvar+1)
    plt.plot(ROC_results[tvar]['scan_steps'], ROC_results[tvar]['dtime_fracs'], label = tvar)
    #plt.plot(ROC_results[tvar]['storage_space_bytes'], ROC_results[tvar]['tprs'], label = tvar)
    
    plt.xlabel('Trigger threshold')
    plt.ylabel('Deadtime fraction')
    plt.grid()
    plt.legend(loc = 'upper right')

plt.show()
```

So, deadtime increases after an energy threshold of around
450 ADC, or once occupancy is lower than 4. 

Some fool is using a float for the energy value! As shown by the 
raw distribution above, hardly any signals will saturate if we 
switch to a less precise data type, like and int16 (see DAQ line 64). 

Repeating everything with this data type actaully keeps deadtime at 
zero for all possible trigger setups. Its always worth triple-checking 
for code optimisations before more complicated solutions.
